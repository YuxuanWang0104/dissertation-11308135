```r
# 初始化结果数据框
total_results_event_rate_shift <- data.frame()

# 运行实验
set.seed(123)
num_simulations <- 200

for (exp in experiments) {
  results <- data.frame(auc = numeric(num_simulations), o_e = numeric(num_simulations), c_slope = numeric(num_simulations), brier_score = numeric(num_simulations))
  
  for (i in 1:num_simulations) {
    set.seed(i)
    
    data_train <- generate_event_rate_shifted_data_1(i, exp$target_mortality_start, exp$target_mortality_end)
    
    # 训练模型
    model <- train_model(data_train)
      
    # 计算性能指标
    metrics <- calculate_performance_metrics(model)
    
    results[i, ] <- unlist(metrics)
  }
  
  # 计算中位数和四分位数性能指标
  median_metrics <- apply(results, 2, median)
  quantile_metrics_25 <- apply(results, 2, quantile, probs = 0.25)
  quantile_metrics_75 <- apply(results, 2, quantile, probs = 0.75)
  
    
  # 格式化结果并保留小数点后7位
  formatted_metrics <- data.frame(
    experiment = exp$name,
    auc_median = median_metrics['auc'],
    auc_25 = quantile_metrics_25['auc'],
    auc_75 = quantile_metrics_75['auc'],
    o_e_median = median_metrics['o_e'],
    o_e_25 = quantile_metrics_25['o_e'],
    o_e_75 = quantile_metrics_75['o_e'],
    c_slope_median = median_metrics['c_slope'],
    c_slope_25 = quantile_metrics_25['c_slope'],
    c_slope_75 = quantile_metrics_75['c_slope'],
    brier_score_median = median_metrics['brier_score'],
    brier_score_25 = quantile_metrics_25['brier_score'],
    brier_score_75 = quantile_metrics_75['brier_score']
  )
  
  total_results_event_rate_shift <- rbind(total_results_event_rate_shift, formatted_metrics)
}

```

```r
library(ggplot2)
library(dplyr)

# 创建对比图的函数
create_comparison_plots <- function(df1, df2, df1_name, df2_name) {
  # 为两个数据框添加来源标识符
  df1 <- df1 %>% mutate(source = df1_name)
  df2 <- df2 %>% mutate(source = df2_name)
  
  # 合并两个数据框
  combined_df <- bind_rows(df1, df2)
  
  # 确保 experiment 列作为因子并按唯一值排序
  combined_df$experiment <- factor(combined_df$experiment, levels = unique(combined_df$experiment))
  
  # 创建误差条图和折线图的函数
  create_error_bar_plot <- function(metric_median, metric_25, metric_75, metric_name, metric_color) {
    ggplot(combined_df, aes(x = experiment, y = !!sym(metric_median), group = source, color = source)) +
      geom_line(size = 1) +
      geom_point(size = 2) +
      geom_errorbar(aes(ymin = !!sym(metric_25), ymax = !!sym(metric_75)), width = 0.2) +
      scale_color_manual(values = c("blue", "red")) +
      theme_minimal() +
      labs(title = paste("Comparison of", metric_name, "between", df1_name, "and", df2_name),
           x = "Experiment", y = metric_name) +
      theme(axis.text.x = element_text(angle = 45, hjust = 1))
  }
  
  # 创建各个性能指标的图表
  plot_auc <- create_error_bar_plot("auc_median", "auc_25", "auc_75", "AUC", c("blue", "red"))
  plot_oe <- create_error_bar_plot("o_e_median", "o_e_25", "o_e_75", "O/E", c("blue", "red"))
  plot_cslope <- create_error_bar_plot("c_slope_median", "c_slope_25", "c_slope_75", "C-Slope", c("blue", "red"))
  plot_brier_score <- create_error_bar_plot("brier_score_median", "brier_score_25", "brier_score_75", "Brier Score", c("blue", "red"))
  
  # 打印图表
  print(plot_auc)
  print(plot_oe)
  print(plot_cslope)
  print(plot_brier_score)
}

# 假设两个数据框分别是 df1 和 df2
df1 <- data.frame(
  experiment = c("gradual_shifted_mortality_1", "gradual_shifted_mortality_2", "gradual_shifted_mortality_3", "gradual_shifted_mortality_4", "gradual_shifted_mortality_5", "gradual_shifted_mortality_6", "gradual_shifted_mortality_7"),
  auc_median = c(0.7078700, 0.7080772, 0.7083887, 0.7080896, 0.7077999, 0.7082595, 0.7080275),
  auc_25 = c(0.7063367, 0.7062108, 0.7066434, 0.7063067, 0.7062709, 0.7065647, 0.7064272),
  auc_75 = c(0.7096219, 0.7096819, 0.7099645, 0.7096910, 0.7094343, 0.7096700, 0.7095164),
  o_e_median = c(0.9881977, 0.8851782, 0.8001696, 0.7289595, 0.6692828, 0.6188507, 0.5759968),
  o_e_25 = c(0.9745920, 0.8699782, 0.7899429, 0.7199494, 0.6621921, 0.6128645, 0.5687778),
  o_e_75 = c(1.0059236, 0.8948727, 0.8126143, 0.7379445, 0.6762084, 0.6257811, 0.5801798),
  c_slope_median = c(1.059122, 1.064393, 1.071248, 1.086389, 1.114673, 1.147113, 1.178710),
  c_slope_25 = c(1.036193, 1.036375, 1.050378, 1.061239, 1.087988, 1.118625, 1.150879),
  c_slope_75 = c(1.086431, 1.085405, 1.091638, 1.110612, 1.137576, 1.175530, 1.213573),
  brier_score_median = c(0.1468590, 0.1475086, 0.1495858, 0.1531115, 0.1579370, 0.1636260, 0.1708512),
  brier_score_25 = c(0.1465506, 0.1472371, 0.1491035, 0.1523683, 0.1570631, 0.1626344, 0.1697112),
  brier_score_75 = c(0.1471223, 0.1479919, 0.1500457, 0.1536213, 0.1586026, 0.1645039, 0.1723447)
)

df2 <- data.frame(
  experiment = c("gradual_shifted_mortality_1", "gradual_shifted_mortality_2", "gradual_shifted_mortality_3", "gradual_shifted_mortality_4", "gradual_shifted_mortality_5", "gradual_shifted_mortality_6", "gradual_shifted_mortality_7"),
  auc_median = c(0.7107952, 0.7107099, 0.7107816, 0.7107894, 0.7106776, 0.7107952, 0.7107201),
  auc_25 = c(0.7101789, 0.7101794, 0.7102417, 0.7102617, 0.7101481, 0.7103526, 0.7102060),
  auc_75 = c(0.7112620, 0.7112433, 0.7112118, 0.7111980, 0.7111808, 0.7112337, 0.7112014),
  o_e_median = c(0.9947271, 0.8899238, 0.8033678, 0.7310040, 0.6712206, 0.6203592, 0.5767001),
  o_e_25 = c(0.9808540, 0.8736501, 0.7935599, 0.7222856, 0.6641199, 0.6135865, 0.5699827),
  o_e_75 = c(1.0134020, 0.9003439, 0.8155676, 0.7414398, 0.6779056, 0.6273584, 0.5814293),
  c_slope_median = c(1.015297, 1.022976, 1.030824, 1.048709, 1.075835, 1.108937, 1.149239),
  c_slope_25 = c(0.9976639, 0.9990947, 1.0086795, 1.0270212, 1.0514233, 1.0836463, 1.1189159),
  c_slope_75 = c(1.041828, 1.041386, 1.050414, 1.074032, 1.096883, 1.130542, 1.176015),
  brier_score_median = c(0.1461497, 0.1469417, 0.1490389, 0.1524725, 0.1572010, 0.1630914, 0.1702532),
  brier_score_25 = c(0.1460842, 0.1467666, 0.1486786, 0.1519202, 0.1565139, 0.1621529, 0.1692556),
  brier_score_75 = c(0.1462659, 0.1471920, 0.1493715, 0.1530911, 0.1579784, 0.1641277, 0.1716324)
)

# 调用函数生成对比图
create_comparison_plots(df1, df2, "Dataset 1", "Dataset 2")
```

```r
# 假设你有一个函数 train_model 用于拟合模型，并返回模型对象
train_model <- function(data) {
  model <- glm(mortality ~ age + sex + chronic_pulmonary_disease + previous_cardiac_surgery + critical_preoperative_state, 
               data = data, family = binomial)
  return(model)
}

# 统计量列表
results <- data.frame()

for (seed in 1:200) {
  data_non_shifted <- generate_event_rate_shifted_data_1(seed, 0.2, 0.25)
  
  # 计算统计量
  stats <- data.frame(
    mean_age = mean(data_non_shifted$age),
    min_age = min(data_non_shifted$age),
    max_age = max(data_non_shifted$age),
    sd_age = sd(data_non_shifted$age),
    sex_proportions_male = prop.table(table(data_non_shifted$sex))["0"],
    sex_proportions_female = prop.table(table(data_non_shifted$sex))["1"],
    cpd_proportions_0 = prop.table(table(data_non_shifted$chronic_pulmonary_disease))[1],
    cpd_proportions_1 = prop.table(table(data_non_shifted$chronic_pulmonary_disease))[2],
    pcs_proportions_0 = prop.table(table(data_non_shifted$previous_cardiac_surgery))[1],
    pcs_proportions_1 = prop.table(table(data_non_shifted$previous_cardiac_surgery))[2],
    cps_proportions_0 = prop.table(table(data_non_shifted$critical_preoperative_state))[1],
    cps_proportions_1 = prop.table(table(data_non_shifted$critical_preoperative_state))[2],
    mortality_proportions_0 = prop.table(table(data_non_shifted$mortality))[1],
    mortality_proportions_1 = prop.table(table(data_non_shifted$mortality))[2]
  )
  
  # 拟合模型并提取系数
  model <- train_model(data_non_shifted)
  coefs <- coef(model)
  
  # 将系数添加到统计量中
  stats <- cbind(stats, t(coefs[c("age", "sex", "chronic_pulmonary_disease", "previous_cardiac_surgery", "critical_preoperative_state")]))
  
  # 将每次循环的统计量结果合并到结果列表中
  results <- rbind(results, stats)
}

# 计算每个统计量的中位数
medians <- apply(results, 2, median)
print(medians)
```
